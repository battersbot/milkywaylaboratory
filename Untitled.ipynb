```python
import aplpy
import astropy
import matplotlib
import matplotlib.pyplot as plt
from astropy.coordinates import SkyCoord
from matplotlib.colors import LogNorm
from astropy.wcs import WCS
from astropy.utils.data import get_pkg_data_filename
from astropy.io import fits
import numpy as np
from astropy.coordinates import match_coordinates_sky
from astropy.convolution import Gaussian2DKernel
from scipy.signal import convolve as scipy_convolve
from astropy.convolution import convolve, convolve_fft
from astropy.convolution import interpolate_replace_nans
```

```python
filename="Xray_Clavel_chandra_pigs-cube.fits"
hdu=fits.open(filename)[0]
#img= hdu.data[50:90, 60:100] * 1e5

hdu.data[0][hdu.data[0] > 1e-9] = np.nan
img_zerod = hdu.data[0].copy()
img_zerod[np.isnan(hdu.data[0])] = 0
kernel = Gaussian2DKernel(15)

scipy_conv = scipy_convolve(hdu.data[0], kernel, mode='same', method='direct')                      
scipy_conv_zerod = scipy_convolve(img_zerod, kernel, mode='same',method='direct')         

astropy_conv0 = convolve(hdu.data[0], kernel)

#smoothed data
plt.figure() 
plt.imshow(hdu.data[0])
plt.contour(astropy_conv0,colors="green")
plt.show()


#Non smoothed data
plt.figure() 
plt.imshow(hdu.data[0])
plt.contour(hdu.data[0],colors="green")
plt.show()
```
